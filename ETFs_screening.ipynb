```python
tickers=[]
desc=[]
def get_tickers(url):
    c=0
    response=requests.get(url)

    html=response.content
    soup = BeautifulSoup(html, 'html.parser')
    tickers_raw=soup.find_all("a", class_="Fw(600) C($linkColor)")
    for elem in tickers_raw:
        tickers.append(str(elem['href']).split('?')[0][7:])
        c=c+1
        print (c,str(elem['href']).split('?')[0][7:])
        desc.append(elem['title'])
    return tickers, desc


urls=["https://finance.yahoo.com/etfs/?offset=0&count=100","https://finance.yahoo.com/etfs/?count=100&offset=100","https://finance.yahoo.com/etfs/?count=100&offset=200", "https://finance.yahoo.com/etfs/?count=100&offset=300","https://finance.yahoo.com/etfs/?count=100&offset=400"] 
    
    
for u in urls:
    d=get_tickers(u)
    
    

```

```python
tickers_f=d[0]
#dic=zip(d[0], d[1])
```

```python
raw=pd.DataFrame()
c=0
for i in tickers_f:
    c=c+1
    print (i,c)
    a=yf.Ticker(i)
    #print (a)
    raw[i]=a.history(period='max')['Close']
```

```python
d1=raw.fillna(value=0)
```

```python
def cal_returns(ticker):
    inx=list(d1[t].index)
    new=[]
    c=False
    a=0
    for elem in range(0,len(d1[ticker])):
       # print (df1[ticker].iloc[elem])
        if d1[ticker].iloc[elem]!=0:
            c=True
            a=a+1
            if c is True and a==1:
                ini=d1[ticker].iloc[elem]
                c=False

 
        
        if d1[ticker].iloc[elem]==0:
            new.append(0)
        elif d1[ticker].iloc[elem-1]==0 and d1[ticker].iloc[elem+1]!=0:
            new.append(0)     
        else:
            new.append(d1[ticker].iloc[elem]/ini*100)
        
    return pd.Series(dict(zip(inx, new)))

#Calculate log returns
def cal_log_returns(ticker):
    inx=list(d1[t].index)
    new=[]
    for elem in range(0,len(d1[ticker])):

  
           
        if d1[ticker].iloc[elem]==0:
            new.append(0)
        elif d1[ticker].iloc[elem-1]==0 and d1[ticker].iloc[elem+1]!=0:
            new.append(0)
        else:
            new.append(np.log(d1[ticker][elem]/d1[ticker][elem-1]))
        
    return pd.Series(dict(zip(inx, new)))

log_ret=pd.DataFrame()
simp_ret=pd.DataFrame()

for t in list(d1.columns):
    print(t)
    log_ret[t]=cal_log_returns(t)
    simp_ret[t]=cal_returns(t)

        
```
