```python
#Calculate simple returns
def cal_returns(ticker):
    inx=list(df1[t].index)
    new=[]
    c=False
    a=0
    for elem in range(0,len(df1[ticker])):
       # print (df1[ticker].iloc[elem])
        if df1[ticker].iloc[elem]!=0:
            c=True
            a=a+1
            if c is True and a==1:
                ini=df1[ticker].iloc[elem]
                c=False

 
        
        if df1[ticker].iloc[elem]==0:
            new.append(0)
        elif df1[ticker].iloc[elem-1]==0 and df1[ticker].iloc[elem+1]!=0:
            new.append(0)     
        else:
            new.append(df1[ticker].iloc[elem]/ini*100)
        
    return pd.Series(dict(zip(inx, new)))

#Calculate log returns
def cal_log_returns(ticker):
    inx=list(df1[t].index)
    new=[]
    for elem in range(0,len(df1[ticker])):

  
           
        if df1[ticker].iloc[elem]==0:
            new.append(0)
        elif df1[ticker].iloc[elem-1]==0 and df1[ticker].iloc[elem+1]!=0:
            new.append(0)
        else:
            new.append(np.log(df1[ticker][elem]/df1[ticker][elem-1]))
        
    return pd.Series(dict(zip(inx, new)))

log_ret=pd.DataFrame()
simp_ret=pd.DataFrame()

for t in list(df1.columns):
    print(t)
    log_ret[t]=cal_log_returns(t)
    simp_ret[t]=cal_returns(t)

```
