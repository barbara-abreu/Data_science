```python
import requests
from bs4 import BeautifulSoup
```

```python
url="https://www.imovirtual.com/en/comprar/apartamento/lisboa/?search%5Bregion_id%5D=11"
response=requests.get(url)
```

```python
html=response.content
soup=BeautifulSoup(html, 'html.parser')
```

```python
#soup.find_all('li',class_="offer-item-price")
lis=soup.find_all('li',class_="offer-item-price")
prices=[str(i.text.strip().strip("€")).replace(" ","") for i in lis]
    
```
